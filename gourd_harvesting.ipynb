```python
from ultralytics import YOLO

# Load base model (Nano is small and fast; use yolov8s.pt for better accuracy)
model = YOLO("yolov8n.pt")

# Train model on your dataset
model.train(
    data="C:/Users/rushi/OneDrive/Desktop/Agriculture Harvesting/project/Gourd_Dataset/data.yaml",  #  path data.yaml
    epochs=50,
    imgsz=640,
    batch=8,
    name="gourd-harvest-detector"
)
```

```python
from ultralytics import YOLO
import cv2

# Load the trained model
model = YOLO("runs/detect/gourd-harvest-detector2/weights/best.pt")

# Load an image
image_path = "C:/Users/rushi/OneDrive/Desktop/Agriculture Harvesting/project/bitter4.jpg"  
results = model(image_path)

# Print results
for result in results:
    boxes = result.boxes
    for box in boxes:
        cls_id = int(box.cls[0].item())
        conf = float(box.conf[0].item())
        label = model.names[cls_id]
        print(f"Detected: {label} (Confidence: {conf:.2f})")

# Show the image with bounding boxes
results[0].show()  # Uses OpenCV
```

```python
import cv2
from ultralytics import YOLO

# Load your best weights
model = YOLO("runs/detect/gourd-harvest-detector2/weights/best.pt")

# Open webcam (0 is usually the default camera)
cap = cv2.VideoCapture("C:/Users/rushi/OneDrive/Desktop/Agriculture Harvesting/project/bg_video1.mp4")

while cap.isOpened():
    success, frame = cap.read()
    if success:
        # Run inference - stream=True is more memory efficient
        results = model.track(frame, persist=True, imgsz=640, conf=0.5)

        # Plot the results on the frame
        annotated_frame = results[0].plot()

        # Display the live feed
        cv2.imshow("Gourd Harvesting Real-Time", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
